```python

import matplotlib.pyplot as plt

# Define o ponto e a reta
P = Vetor([1, 3])
l = Reta(ponto=Vetor([1, 1]), normal=Vetor([3, 4]))

# Cria figura e plota a reta
fig = l.plot(cor='gray', limite=5)

# Plota o ponto
P.plot(fig=fig, cor='r')

# Mostra o gráfico
plt.title("Ponto e reta no plano")
plt.show()
```

```python
import matplotlib.pyplot as plt


# Define duas retas no plano
r1 = Reta(ponto=Vetor([0, 1]), normal=Vetor([2, -1]))  # reta 1
r2 = Reta(ponto=Vetor([1, 0]), normal=Vetor([1, 2]))   # reta 2

# Calcula ponto de interseção
p_intersec = r1.interseccao(r2)

# Plota as duas retas
fig = r1.plot(cor='b', limite=5)
fig = r2.plot(fig=fig, cor='g', limite=5)

# Plota o ponto de interseção
ax = fig.gca()
ax.plot(p_intersec.x, p_intersec.y, 'ro', label='Interseção')
ax.legend()

plt.title("Interseção de duas retas no plano")
plt.show()
```

```python
import matplotlib.pyplot as plt
from modules.Vetor import Vetor
from modules.Reta import Reta
from modules.Circulo import Circulo

# Define a reta e o círculo
reta = Reta(ponto=Vetor([10, 0]), diretor=Vetor([-1, 1]))
circulo = Circulo(centro=Vetor([2, 1]), raio=2)

# Calcula os pontos de interseção
pontos = reta.interseccao(circulo)

# Plota o círculo
fig = circulo.plot(cor='blue', limite=4)

# Plota a reta na mesma figura
fig = reta.plot(fig=fig, cor='gray', limite=4)

# Plota os pontos de interseção
ax = fig.gca()
for p in pontos:
    ax.plot(p.x, p.y, 'ro', label='Interseção')

# Ajustes finais
if pontos:
    ax.legend()
plt.title("Interseção entre reta e círculo")
plt.show()
```
